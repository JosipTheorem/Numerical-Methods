To keep $0 \le x \le \pi$ in the series we can do the following:

If we have $ x > \pi$, we can write it as $x = \tilde{x} + n \pi$, where n=1,2,3,... and $0 \le \tilde{x} \le \pi$.

Now $sin(x)=sin(\tilde{x} + n \pi)=sin(\tilde{x})cos(n\pi)+sin(n\pi)cos(\tilde{x})=(-1)^nsin(\tilde{x})$.

Where $sin(\tilde{x})$ is callculated with the series.



```python
###Series to keep 0 ‚â§ ùë• ‚â§ùúã  

def Taylor_sinx_4(x,N):
    m=int(np.floor(x/np.pi))
    x=x-m*np.pi
    S=0
    for n in range(1,N+1):
        
        if ((abs((-1)**(n-1) * x**(2*n-1)))/(np.math.factorial(2*n-1)) <=10e-7 * abs(S)):
            
            break
        
        S=((-1)**(n-1) * x**(2*n-1))/(np.math.factorial(2*n-1))+S
        
        
    return S * (-1)**m
```

### The equation of motion:

### $ F_k(x)+F_{ext}(x,t)=m\frac{d^2x}{dt^2}$

### Potential:

### $V(x) = \frac{1}{p}kx^p$

### Force:

### $F_k(x)=-\frac{dV(x)}{dx}=-kx^{p-1}$

### The equation of motion becomes:

### $m\frac{d^2x}{dt^2}=F_{ext}(x,t)-kx^{p-1}$

### External force will be omited for simplicity

```python
k = 1
m = 1
x_0 = 1
v_0 = 0
w_0 = np.sqrt(k/m)
```

```python
ti = 0
#tf = 5*(2*np.pi/w_0)
tf=80
N = 1000000
```

```python
#RK4
h = (tf-ti)/N
t = np.arange(ti,tf,h)

def RK4(p):
    
    def F(y, u, x):
        return -(w_0**2)* y**(p-1)


    
    

    
    x_t = []
    v_t = []

    y = x_0
    u = v_0 

    for x in t:
        x_t.append(y)
        v_t.append(u)

        m1 = h*u
        k1 = h*F(y, u, x)  #(x, v, t)

        m2 = h*(u + 0.5*k1)
        k2 = h*F(y+0.5*m1, u+0.5*k1, x+0.5*h)

        m3 = h*(u + 0.5*k2)
        k3 = h*F(y+0.5*m2, u+0.5*k2, x+0.5*h)

        m4 = h*(u + k3)
        k4 = h*F(y+m3, u+k3, x+h)

        y += (m1 + 2*m2 + 2*m3 + m4)/6
        u += (k1 + 2*k2 + 2*k3 + k4)/6
        
    return x_t,v_t
```

```python
p=4 
plt.plot(t,RK4(p)[0],label='x_t')
plt.plot(t,RK4(p)[1],label='v_t')
plt.axhline(y=0,linewidth=2,ls='--',color='red')
plt.title('p=%d'%p,fontsize='16')

plt.xlabel('t',fontsize='13') 
plt.ylabel(r'$x (t)/v(t)$',fontsize='13')
plt.legend(fontsize=13)

plt.show()
```

The shapes of the oscillations change with p because the force changes with p and therefor equation of motion changes with p and therefor different solution satisfies the equation of motion.

```python
x_0 = 1
##### Algorith for finding the period
def period_finder(p,x1):
    T=[]
   
    b = RK4(p)[0]
    for i in range(0,len(b)-1):
        if (b[i]<x1-0.000001 and b[i+1]>x1-0.000001):
            T.append(t[i])
            
    return T
    
```

Algorithm is finding values of $t$ for wich $x[t]\approx x_0$.

For $p=6$ and $x_0=1$ period is $\approx $ 8.41312, wich can be verified on the first graph above.

```python
avg=[]
x_0 = 1
for i in range(0,5):
    
    T=period_finder(6,x_0)
    d=[]
    for j in range (0,len(T)-2):
        d.append(T[j+1]-T[j])
    avg.append(sum(d)/len(d))
    x_0 = x_0 + 0.5

x_0 = 1
A=[]
for i in range(0,5):
  
    A.append(x_0)
    x_0 = x_0 + 0.5
```
