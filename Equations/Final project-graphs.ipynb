# 1. TRIAL-AND-ERROR SEARCH GRAPHS

```python
def box_potential(x):
    
    return np.sqrt(V_0-x)*np.tan(np.sqrt(V_0-x))-np.sqrt(x)
```

# $V_0$ = 10

```python
V_0 = 10
```

```python
xs=np.linspace(0,0.1,1000000)
plt.plot(xs,box_potential(xs))

plt.title('E_B = [0 , 0.1] ',fontsize='16')

plt.xlabel('E_B',fontsize='13') 
```

```python
xs=np.linspace(8,10,1000000)
plt.plot(xs,box_potential(xs))

plt.title('E_B = [8 , 10] ',fontsize='16')

plt.xlabel('E_B',fontsize='13') 
```
